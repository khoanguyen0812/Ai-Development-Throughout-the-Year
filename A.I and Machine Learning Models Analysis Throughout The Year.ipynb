## A.I And Machine Learning Model Developments Analysis.

- This dataset looks at the development of each A.I and Machine Learning reported each years by different sources. This is helps to tell us how much resources are needed and how accepting the public is of them.

    Year: The year when the AI model was developed.
        Measured as: Numerical (e.g., 2016, 2017, ... 2022).

    Month: The month when the AI model was developed.
        Measured as: Numerical (1-12).

    Model_Name: The name of the AI model.
        Measured as: Text (e.g., Model_17).

    Models_Created: The number of AI models created.
        Measured as: Numerical (e.g., 1, 2, 3, ...).

    Algorithm_Type: The type of AI algorithm used (e.g., Reinforcement, Unsupervised).
        Measured as: Categorical (e.g., "Reinforcement", "Unsupervised").

    Data_Size_GB: The size of the data used, in gigabytes.
        Measured as: Numerical (e.g., 184.95 GB).

    Training_Time_Hours: The time taken for training the model, in hours.
        Measured as: Numerical (e.g., 90.05 hours).

    Accuracy: The accuracy metric of the model.
        Measured as: Numerical, typically a fraction or percentage (e.g., 0.95).

    Precision: The precision metric of the model.
        Measured as: Numerical, typically a fraction or percentage (e.g., 0.88).

    Recall: The recall metric of the model.
        Measured as: Numerical, typically a fraction or percentage (e.g., 1.00).

    F1_Score: The F1 score metric of the model, combining precision and recall.
        Measured as: Numerical, typically a fraction or percentage (e.g., 0.59).

    Validation_Split: The proportion of the dataset used for validation.
        Measured as: Numerical, typically a fraction (e.g., 0.21).

    Epochs: The number of training epochs.
        Measured as: Numerical (e.g., 498).

    Batch_Size: The size of the batch used in training.
        Measured as: Numerical (e.g., 128).

    Optimizer: The optimization algorithm used for training.
        Measured as: Categorical (e.g., "RMSprop", "Adamax").

    Investment_Million_USD: The investment in AI development, in million USD.
        Measured as: Numerical (e.g., 3309.28 million USD).

    Public_Acceptance: The initial representation of public acceptance.
        Measured as: Numerical, typically a fraction or percentage (e.g., 0.68).

    Investment_in_Billion_USD: The investment represented in billion USD.
        Measured as: Numerical (e.g., 3.309 billion USD).

    Public_Acceptance_Percentage: Public acceptance represented as a percentage.
        Measured as: Numerical, percentage (e.g., 59.30%).

    Number_of_Employees: Estimated number of employees working on AI projects.
        Measured as: Numerical (e.g., 100).

    Tech_Advancements: Level of technological advancements.
        Measured as: Categorical (e.g., "Low", "Moderate", "High", "Very High").

    Industry_Adoption: Rate at which AI technologies are adopted across sectors.
        Measured as: Categorical (e.g., "Low", "Moderate", "High").

    Regulatory_Environment: The regulatory environment affecting AI development.
        Measured as: Categorical (e.g., "Flexible", "Moderate", "Strict", "Very Strict").

    Public_Awareness: General public's understanding and awareness of AI technologies.
        Measured as: Categorical (e.g., "Low", "Moderate", "High", "Very High").

Questions we are trying to answer:
- How accepting is the public of A.I and Machine Learning models over the years?

- What are some factors that are included in A.Is and Machine Learning Models that are successful?

- If a start-up tech company wishes to get involved in A.I development, how much capital and resources should they put in going foward from 2022?

```python
def check_dataframe_values(df):
    # Check for missing values in the entire DataFrame
    if ai_dev.isnull().any().any():
        print("Missing values found")
        missing_values_count = df.isnull().sum().sum()
        print(f"Total missing values: {missing_values_count}")
    
    # Check for white spaces in string (object) columns
    white_spaces_columns = [col for col in ai_dev.columns if ai_dev[col].dtype == 'object' and df[col].str.isspace().any()]
    if white_spaces_columns:
        print("White spaces found in columns:", white_spaces_columns)
    
    # Check for duplicate rows
    if ai_dev.duplicated().any():
        print("Duplicate rows found")
        duplicate_rows_count = df.duplicated().sum()
        print(f"Total duplicate rows: {duplicate_rows_count}")

# Now, let's use the adjusted function on the dataset
check_dataframe_values(ai_dev)
```

```python
# Performing KNN imputation on qualitative, mode imputation for quantitative data

from sklearn.impute import KNNImputer, SimpleImputer

# Separating the dataset into quantitative and qualitative parts
quantitative_columns = ai_dev.select_dtypes(include=['float64', 'int64']).columns
qualitative_columns = ai_dev.select_dtypes(include=['object']).columns

# Quantitative data for KNN imputation
quantitative_data = ai_dev[quantitative_columns]

# KNN Imputer for quantitative variables
knn_imputer = KNNImputer(n_neighbors=5)
quantitative_data_imputed = knn_imputer.fit_transform(quantitative_data)

# Convert the imputed array back to a pandas DataFrame
quantitative_data_imputed_df = pd.DataFrame(quantitative_data_imputed, columns=quantitative_columns)

# Now, for qualitative variables, we'll use a for loop with SimpleImputer
mode_imputer = SimpleImputer(strategy='most_frequent')  # Initialize the SimpleImputer for mode imputation

# Iterate over qualitative columns and apply mode imputation
for column in qualitative_columns:
    # Impute the missing values in the column
    imputed_data = mode_imputer.fit_transform(ai_dev[[column]])
    
    # Update the DataFrame with imputed data - Ensure it's a 1D array for the column
    ai_dev[column] = imputed_data.ravel()  # .ravel() converts the 2D array to a 1D array

# Since we've already updated ai_dev directly for qualitative columns, let's update it for quantitative columns as well
ai_dev.update(quantitative_data_imputed_df)

# Verify if there are any missing values left in the entire DataFrame
missing_values_after_imputation = ai_dev.isnull().sum()

print(missing_values_after_imputation)

print(ai_dev.isnull().sum())
```

#### Question 1: How accepting is the public of A.I and Machine Learning models over the years?

```python
# Define a list of dictionaries for each mapping operation
mappings = [
    {
        'original_column': 'Tech_Advancements',
        'mapped_column': 'Tech_Advancement_Mapped',
        'mapping_dict': {'Low': 1, 'Moderate': 2, 'High': 3, 'Very High': 4}
    },
    {
        'original_column': 'Industry_Adoption',
        'mapped_column': 'Industry_Adoption_Mapped',
        'mapping_dict': {'Low': 1, 'Moderate': 2, 'High': 3}
    },
    {
        'original_column': 'Regulatory_Environment',
        'mapped_column': 'Regulatory_Environment_Mapped',
        'mapping_dict': {'Flexible': 1, 'Moderate': 2, 'Strict': 3, 'Very Strict': 4}
    },
    {
        'original_column': 'Public_Awareness',
        'mapped_column': 'Public_Awareness_Mapped',
        'mapping_dict': {'Low': 1, 'Moderate': 2, 'High': 3, 'Very High' : 4}
    }
]

# Iterate over each mapping operation
for mapping in mappings:
    # Apply the mapping
    ai_dev[mapping['mapped_column']] = ai_dev[mapping['original_column']].map(mapping['mapping_dict'])
    
    # Find the index of the original column
    original_index = ai_dev.columns.tolist().index(mapping['original_column'])
    
    # Reorder columns to place the mapped column immediately after the original column
    cols = ai_dev.columns.tolist()
    # Insert the mapped column name right after the original column's position
    cols.insert(original_index + 1, cols.pop(cols.index(mapping['mapped_column'])))
    # Apply the new column order to the dataframe
    ai_dev = ai_dev[cols]

ai_dev.dtypes
```

```python
# Define a list of columns to transform and their mappings
transform_columns = ['Algorithm_Type', 'Optimizer']
transform_mappings = {
    'Algorithm_Type': {'Unsupervised': 1, 'Reinforcement': 2, 'Supervised': 3},
    'Optimizer': {'RMSprop': 1, 'Adamax': 2, 'Adam': 3, 'SGD': 4}
}

# Iterate over the columns and apply the mappings using a for loop
for column in transform_columns:
    # Create a new column name for the transformed values
    transformed_column_name = f"{column}_Mapped"
    # Apply the mapping
    ai_dev[transformed_column_name] = ai_dev[column].map(transform_mappings[column])
    
    # Find the index of the original column
    original_index = ai_dev.columns.tolist().index(column)
    
    # Reorder columns to place the mapped column immediately after the original column
    cols = ai_dev.columns.tolist()
    # Insert the mapped column name right after the original column's position
    cols.insert(original_index + 1, cols.pop(cols.index(transformed_column_name)))
    # Apply the new column order to the dataframe
    ai_dev = ai_dev[cols]

# Display the datatypes of the updated dataframe to confirm the changes
ai_dev.dtypes
```

```python
# Understanding public acceptance of A.I and Machine Learning within a 5 year interval

# Filter the dataset for years 2017 and 2022 to compare public acceptance
filtered_data = ai_dev[ai_dev['Year'].isin([2017, 2022])]

plt.figure(figsize=(12, 8))

# Create the violin plot without specifying 'inner' to avoid clutter
sns.violinplot(x="Year", y="Public_Acceptance", hue="Algorithm_Type", data=filtered_data, palette="muted")

# Overlay a point plot to show the median
# Use the 'dodge' parameter to align median points with violins
sns.pointplot(x="Year", y="Public_Acceptance", hue="Algorithm_Type", data=filtered_data,
              dodge=True, join=False, markers="D", capsize=0.1, color="black", estimator=np.median)

plt.title('Public Acceptance of Different Algorithm Types in 2017 vs 2022', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Public Acceptance', fontsize=12)
plt.xticks(rotation=45)

# Adjusting legend to ensure it's clear and informative
handles, labels = plt.gca().get_legend_handles_labels()
# Deduplicate legend entries (violinplot and pointplot create duplicate legends)
half = len(handles) // 2
plt.legend(handles[:half], labels[:half], title='Algorithm Type', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```
